# Determining minimum effect sizes for each statisical test.

<br>
<br>

### H1 and H4. Independent T-Test between children (run 1 or run 2) and adults across ROIs.

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.power import FTestAnovaPower, TTestIndPower, TTestPower

# getting minimum effect size needed for 80% power with a fixed sample size (different for each test)
power = 0.8
alpha = 0.05

# possible effect sizes
effect_size = np.linspace(0.2, 1.2, 50)
```

```python
# observations for this test
nobs = np.array([56])
power_analysis = TTestIndPower()
result = power_analysis.solve_power(power=power, nobs1=nobs[0], alpha=alpha)

# plot the range of effect size possible
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1, 1,1)
title = f'For an Independent t-test between children and adults \n an effect size={result:0.3f} will be sufficiently powered'
fig = power_analysis.plot_power(dep_var='effect_size',
                                 nobs = nobs,
                                 effect_size = effect_size,
                                 alpha = alpha,
                                 ax=ax, 
                                 title=title )
plt.axhline(power, color='g', linestyle='--')
plt.axvline(result, color='red')
None
```

### H2. One-factor ANOVA between fast and slow regions in children.

```python
# observations for this test
nobs = np.array([26])
power_analysis = FTestAnovaPower()
result = power_analysis.solve_power(power=power, nobs=nobs[0], alpha=alpha)

# plot the range of effect size possible
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1, 1, 1)
title = f'For an Independent t-test between children and adults \n an effect size={result:0.3f} will be sufficiently powered'
fig = power_analysis.plot_power(dep_var='effect_size',
                                 nobs = nobs,
                                 effect_size = effect_size,
                                 alpha = alpha,
                                 ax=ax, 
                                 title=title )
plt.axhline(power, color='g', linestyle='--')
plt.axvline(result, color='red')
None
```

### H3. Dependent samples T-Test within each ROI across childrens' runs

```python
# observations for this test
nobs = np.array([26])
power_analysis = TTestPower()
result = power_analysis.solve_power(power=power, nobs=nobs[0], alpha=alpha)

# plot the range of effect size possible
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1, 1,1)
title = f'For an Independent t-test between children and adults \n an effect size={result:0.3f} will be sufficiently powered'
fig = power_analysis.plot_power(dep_var='effect_size',
                                 nobs = nobs,
                                 effect_size = effect_size,
                                 alpha = alpha,
                                 ax=ax, 
                                 title=title )
plt.axhline(power, color='g', linestyle='--')
plt.axvline(result, color='red')
None
```
